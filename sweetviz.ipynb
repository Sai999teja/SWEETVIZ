```python
import sweetviz
```

```python
import pandas as pd
data = pd.read_csv('train (4).csv')
```

```python
data.head()
```

```python
report = sweetviz.analyze([data,'train'],target_feat = 'SalePrice')      #it will take a while
```

```python
report.show_html('myreport.html')
```

```python
data2 = pd.read_csv('train (1).csv')
```

```python
data2.head()
```

```python
report2 = sweetviz.analyze([data2,'train'],target_feat = 'Survived')  
```

```python
report2.show_html('myreport2.html')
```

```python
#we can also compare with other data frames with similar features like train and test
```

```python
comp_report = sweetviz.compare([data2,'Train'],[test,'Test'],target_feat = 'Survived')
```
